## **05. Modelos de Deep Learning para Pronóstico de Heladas**

# Este notebook implementa y evalúa modelos de Deep Learning:
# 1. **MLP (Multi-Layer Perceptron)** - Red neuronal feedforward
# 2. **LSTM (Long Short-Term Memory)** - Red neuronal recurrente
# 3. **CNN-1D (Convolutional Neural Network)** - Red convolucional 1D
# 
# **Ventajas del Deep Learning:**
# - Capturan patrones temporales complejos
# - Aprenden representaciones automáticamente
# - Manejan dependencias de largo plazo
# - Escalables a grandes volúmenes de datos

```python

# %% [markdown]
# ## 1. Configuración inicial y preparación del entorno

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Deep Learning
import tensorflow as tf
import keras
import tensorflow
import tensorflow.keras as keras   

from tensorflow import keras
from tensorflow.keras import layers, models, callbacks, optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.layers import Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import Adam

# Utilidades
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import joblib
from tqdm import tqdm
import json
import gc

# Configuración
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 11

# Verificar GPU
print("TensorFlow version:", tf.__version__)
print("GPU disponible:", len(tf.config.list_physical_devices('GPU')) > 0)
if len(tf.config.list_physical_devices('GPU')) > 0:
    print("GPU:", tf.config.list_physical_devices('GPU')[0])

# Semilla para reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)

# %%
# Cargar datos procesados
print("Cargando datos procesados...")
train_data = pd.read_csv('../data/processed/train_data.csv', parse_dates=['date'])
val_data = pd.read_csv('../data/processed/val_data.csv', parse_dates=['date'])
test_data = pd.read_csv('../data/processed/test_data.csv', parse_dates=['date'])

print(f"Train: {len(train_data):,} registros")
print(f"Val: {len(val_data):,} registros")
print(f"Test: {len(test_data):,} registros")

# %% [markdown]
# ## 2. Preparación de datos para Deep Learning

# %%
# Definir features (mismo conjunto que ML)
feature_columns = [
    'tmax', 'tmean', 'rh', 'pressure', 'wind', 'precipitation',
    'temp_range', 'temp_deviation',
    'month', 'day_of_year', 'week_of_year',
    'season_winter', 'season_spring', 'season_summer', 'season_autumn',
    'tmin_lag1', 'tmin_lag2', 'tmin_lag3', 'tmin_lag7',
    'tmax_lag1', 'tmean_lag1',
    'tmin_ma3', 'tmin_ma7', 'tmean_ma3', 'tmean_ma7',
    'tmin_std3', 'tmin_std7'
]

# Verificar disponibilidad
available_features = [col for col in feature_columns if col in train_data.columns]
feature_columns = available_features
print(f"Features disponibles: {len(feature_columns)}")

# %%
# Función para crear secuencias para modelos temporales (LSTM, CNN)
def create_sequences(data, feature_cols, target_col, sequence_length=7):
    """
    Crea secuencias para modelos temporales
    """
    sequences = []
    targets = []
    
    # Agrupar por estación para mantener continuidad temporal
    for station in data['station'].unique():
        station_data = data[data['station'] == station].sort_values('date')
        
        # Features y target
        X_station = station_data[feature_cols].values
        y_station = station_data[target_col].values
        
        # Crear secuencias
        for i in range(sequence_length, len(station_data)):
            sequences.append(X_station[i-sequence_length:i])
            targets.append(y_station[i])
    
    return np.array(sequences), np.array(targets)

# %%
# Preparar datos para diferentes arquitecturas

# 1. Para MLP: datos tabulares simples
def prepare_mlp_data(df, feature_cols, target_col='tmin'):
    df_clean = df.dropna(subset=feature_cols + [target_col])
    X = df_clean[feature_cols].values
    y = df_clean[target_col].values
    info = {
        'dates': df_clean['date'].values,
        'stations': df_clean['station'].values if 'station' in df_clean.columns else None
    }
    return X, y, info

# 2. Para LSTM/CNN: secuencias temporales
SEQUENCE_LENGTH = 7  # Usar últimos 7 días

print("Preparando datos...")

# MLP
X_train_mlp, y_train_mlp, train_info = prepare_mlp_data(train_data, feature_columns)
X_val_mlp, y_val_mlp, val_info = prepare_mlp_data(val_data, feature_columns)
X_test_mlp, y_test_mlp, test_info = prepare_mlp_data(test_data, feature_columns)

# LSTM/CNN
print(f"Creando secuencias de longitud {SEQUENCE_LENGTH}...")
X_train_seq, y_train_seq = create_sequences(train_data, feature_columns, 'tmin', SEQUENCE_LENGTH)
X_val_seq, y_val_seq = create_sequences(val_data, feature_columns, 'tmin', SEQUENCE_LENGTH)
X_test_seq, y_test_seq = create_sequences(test_data, feature_columns, 'tmin', SEQUENCE_LENGTH)

print(f"\nFormas de datos:")
print(f"MLP - Train: X={X_train_mlp.shape}, y={y_train_mlp.shape}")
print(f"LSTM/CNN - Train: X={X_train_seq.shape}, y={y_train_seq.shape}")

# %%
# Escalado de datos (crítico para redes neuronales)
print("Escalando features...")

# Usar StandardScaler para features
scaler_X = StandardScaler()
X_train_mlp_scaled = scaler_X.fit_transform(X_train_mlp)
X_val_mlp_scaled = scaler_X.transform(X_val_mlp)
X_test_mlp_scaled = scaler_X.transform(X_test_mlp)

# Para secuencias, escalar manteniendo la estructura 3D
X_train_seq_scaled = X_train_seq.copy()
X_val_seq_scaled = X_val_seq.copy()
X_test_seq_scaled = X_test_seq.copy()

for i in range(X_train_seq.shape[2]):  # Para cada feature
    # Calcular media y std del training
    feature_data = X_train_seq[:, :, i].flatten()
    mean = feature_data.mean()
    std = feature_data.std()
    
    # Aplicar escalado
    X_train_seq_scaled[:, :, i] = (X_train_seq[:, :, i] - mean) / std
    X_val_seq_scaled[:, :, i] = (X_val_seq[:, :, i] - mean) / std
    X_test_seq_scaled[:, :, i] = (X_test_seq[:, :, i] - mean) / std

# Escalar target (opcional, pero puede ayudar)
scaler_y = StandardScaler()
y_train_mlp_scaled = scaler_y.fit_transform(y_train_mlp.reshape(-1, 1)).flatten()
y_val_mlp_scaled = scaler_y.transform(y_val_mlp.reshape(-1, 1)).flatten()
y_train_seq_scaled = scaler_y.transform(y_train_seq.reshape(-1, 1)).flatten()
y_val_seq_scaled = scaler_y.transform(y_val_seq.reshape(-1, 1)).flatten()

# %% [markdown]
# ## 3. Funciones auxiliares

# %%
def evaluate_dl_model(y_true, y_pred, model_name="Model", history=None):
    """
    Evalúa un modelo de Deep Learning
    """
    # Métricas básicas
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    # Métricas de clasificación (heladas)
    frost_true = (y_true <= 0).astype(int)
    frost_pred = (y_pred <= 0).astype(int)
    
    if frost_true.sum() > 0:
        precision = precision_score(frost_true, frost_pred, zero_division=0)
        recall = recall_score(frost_true, frost_pred, zero_division=0)
        f1 = f1_score(frost_true, frost_pred, zero_division=0)
        
        cm = confusion_matrix(frost_true, frost_pred)
        tn, fp, fn, tp = cm.ravel()
        tss = (tp/(tp+fn)) - (fp/(fp+tn)) if (tp+fn) > 0 and (fp+tn) > 0 else 0
    else:
        precision = recall = f1 = tss = 0
    
    metrics = {
        'model': model_name,
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'tss': tss
    }
    
    # Visualización
    fig = plt.figure(figsize=(15, 10))
    
    # Si hay history, mostrar curvas de entrenamiento
    if history is not None:
        # 2x3 grid
        ax1 = plt.subplot(2, 3, 1)
        ax2 = plt.subplot(2, 3, 2)
        ax3 = plt.subplot(2, 3, 3)
        ax4 = plt.subplot(2, 3, 4)
        ax5 = plt.subplot(2, 3, 5)
        ax6 = plt.subplot(2, 3, 6)
        
        # Training curves
        ax1.plot(history.history['loss'], label='Train')
        ax1.plot(history.history['val_loss'], label='Validation')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.set_title('Training History')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
    else:
        # 2x2 grid
        ax2 = plt.subplot(2, 2, 1)
        ax3 = plt.subplot(2, 2, 2)
        ax4 = plt.subplot(2, 2, 3)
        ax6 = plt.subplot(2, 2, 4)
    
    # Scatter plot
    ax2.scatter(y_true, y_pred, alpha=0.5, s=10)
    ax2.plot([-20, 20], [-20, 20], 'r--', lw=2)
    ax2.set_xlabel('Temperatura Real (°C)')
    ax2.set_ylabel('Temperatura Predicha (°C)')
    ax2.set_title(f'{model_name}: Predicho vs Real')
    ax2.text(0.05, 0.95, f'R² = {r2:.3f}\nRMSE = {rmse:.3f}°C', 
             transform=ax2.transAxes, verticalalignment='top',
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    ax2.grid(True, alpha=0.3)
    
    # Error distribution
    errors = y_true - y_pred
    ax3.hist(errors, bins=50, alpha=0.7, edgecolor='black')
    ax3.axvline(x=0, color='red', linestyle='--')
    ax3.set_xlabel('Error (°C)')
    ax3.set_ylabel('Frecuencia')
    ax3.set_title('Distribución de Errores')
    ax3.grid(True, alpha=0.3)
    
    # Error vs temperature
    ax4.scatter(y_true, errors, alpha=0.5, s=10)
    ax4.axhline(y=0, color='red', linestyle='--')
    ax4.set_xlabel('Temperatura Real (°C)')
    ax4.set_ylabel('Error (°C)')
    ax4.set_title('Error vs Temperatura')
    ax4.grid(True, alpha=0.3)
    
    # Learning rate schedule if available
    if history is not None and 'lr' in history.history:
        ax5.plot(history.history['lr'])
        ax5.set_xlabel('Epoch')
        ax5.set_ylabel('Learning Rate')
        ax5.set_title('Learning Rate Schedule')
        ax5.set_yscale('log')
        ax5.grid(True, alpha=0.3)
    
    # Confusion matrix
    if frost_true.sum() > 0:
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=['No Helada', 'Helada'],
                    yticklabels=['No Helada', 'Helada'],
                    ax=ax6)
        ax6.set_title(f'Matriz de Confusión\nTSS = {tss:.3f}')
    
    plt.suptitle(f'Análisis de Desempeño - {model_name}', fontsize=14)
    plt.tight_layout()
    plt.savefig(f'../results/figures/{model_name.lower().replace(" ", "_")}_analysis.png', dpi=300)
    plt.show()
    
    return metrics

# %% [markdown]
# ## 4. Modelo 1: Multi-Layer Perceptron (MLP)

# %%
print("="*70)
print("MODELO 1: MULTI-LAYER PERCEPTRON (MLP)")
print("="*70)

# Arquitectura MLP
def create_mlp_model(input_dim, learning_rate=0.001):
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        
        Dense(32, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        
        Dense(16, activation='relu'),
        
        Dense(1)  # Salida para regresión
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='mse',
        metrics=['mae']
    )
    
    return model

# Crear modelo
mlp_model = create_mlp_model(X_train_mlp_scaled.shape[1])
mlp_model.summary()

# %%
# Callbacks
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=20,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-6,
    verbose=1
)

checkpoint = ModelCheckpoint(
    '../models/trained/mlp_best.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=0
)

# Entrenar
print("\nEntrenando MLP...")
history_mlp = mlp_model.fit(
    X_train_mlp_scaled, y_train_mlp_scaled,
    validation_data=(X_val_mlp_scaled, y_val_mlp_scaled),
    epochs=100,
    batch_size=256,
    callbacks=[early_stop, reduce_lr, checkpoint],
    verbose=1
)

# %%
# Predicciones y evaluación
y_pred_mlp_scaled = mlp_model.predict(X_test_mlp_scaled).flatten()
y_pred_mlp = scaler_y.inverse_transform(y_pred_mlp_scaled.reshape(-1, 1)).flatten()

# Evaluar
mlp_metrics = evaluate_dl_model(y_test_mlp, y_pred_mlp, "MLP", history_mlp)

print(f"\nMétricas MLP:")
print(f"RMSE: {mlp_metrics['rmse']:.3f}°C")
print(f"MAE: {mlp_metrics['mae']:.3f}°C")
print(f"R²: {mlp_metrics['r2']:.3f}")
print(f"TSS: {mlp_metrics['tss']:.3f}")

# %% [markdown]
# ## 5. Modelo 2: LSTM (Long Short-Term Memory)

# %%
print("\n" + "="*70)
print("MODELO 2: LSTM")
print("="*70)

# Arquitectura LSTM
def create_lstm_model(sequence_length, n_features, learning_rate=0.001):
    model = Sequential([
        Input(shape=(sequence_length, n_features)),
        
        LSTM(64, return_sequences=True),
        Dropout(0.2),
        
        LSTM(32, return_sequences=False),
        Dropout(0.2),
        
        Dense(16, activation='relu'),
        Dropout(0.1),
        
        Dense(1)
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='mse',
        metrics=['mae']
    )
    
    return model

# Crear modelo
lstm_model = create_lstm_model(SEQUENCE_LENGTH, X_train_seq_scaled.shape[2])
lstm_model.summary()

# %%
# Entrenar LSTM
print("\nEntrenando LSTM...")

# Callbacks específicos para LSTM
early_stop_lstm = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

checkpoint_lstm = ModelCheckpoint(
    '../models/trained/lstm_best.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=0
)

# Convierte a float32 (o float64)
X_train_seq_scaled = X_train_seq_scaled.astype(np.float32)
y_train_seq_scaled = y_train_seq_scaled.astype(np.float32)
X_val_seq_scaled = X_val_seq_scaled.astype(np.float32)
y_val_seq_scaled = y_val_seq_scaled.astype(np.float32)

# Ahora sí puedes entrenar sin el error:
history_lstm = lstm_model.fit(
    X_train_seq_scaled, y_train_seq_scaled,
    validation_data=(X_val_seq_scaled, y_val_seq_scaled),
    epochs=50,
    batch_size=128,
    callbacks=[early_stop_lstm, reduce_lr, checkpoint_lstm],
    verbose=1
)


# %%
# Convierte a float32 antes de predecir
X_test_seq_scaled = X_test_seq_scaled.astype(np.float32)

# Predicciones y evaluación LSTM
y_pred_lstm_scaled = lstm_model.predict(X_test_seq_scaled).flatten()
y_pred_lstm = scaler_y.inverse_transform(y_pred_lstm_scaled.reshape(-1, 1)).flatten()

# Evaluar
lstm_metrics = evaluate_dl_model(y_test_seq, y_pred_lstm, "LSTM", history_lstm)

print(f"\nMétricas LSTM:")
print(f"RMSE: {lstm_metrics['rmse']:.3f}°C")
print(f"MAE: {lstm_metrics['mae']:.3f}°C")
print(f"R²: {lstm_metrics['r2']:.3f}")
print(f"TSS: {lstm_metrics['tss']:.3f}")

# %% [markdown]
# ## 6. Modelo 3: CNN-1D (Convolutional Neural Network 1D)

# %%
print("\n" + "="*70)
print("MODELO 3: CNN-1D")
print("="*70)

# Arquitectura CNN-1D
def create_cnn1d_model(sequence_length, n_features, learning_rate=0.001):
    model = Sequential([
        Input(shape=(sequence_length, n_features)),
        
        Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        
        Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'),
        BatchNormalization(),
        Dropout(0.2),
        
        Conv1D(filters=16, kernel_size=3, activation='relu', padding='same'),
        
        Flatten(),
        
        Dense(32, activation='relu'),
        Dropout(0.2),
        
        Dense(1)
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='mse',
        metrics=['mae']
    )
    
    return model

# Crear modelo
cnn_model = create_cnn1d_model(SEQUENCE_LENGTH, X_train_seq_scaled.shape[2])
cnn_model.summary()

# %%
# Entrenar CNN
print("\nEntrenando CNN-1D...")

checkpoint_cnn = ModelCheckpoint(
    '../models/trained/cnn1d_best.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=0
)

history_cnn = cnn_model.fit(
    X_train_seq_scaled, y_train_seq_scaled,
    validation_data=(X_val_seq_scaled, y_val_seq_scaled),
    epochs=50,
    batch_size=128,
    callbacks=[early_stop, reduce_lr, checkpoint_cnn],
    verbose=1
)

# %%
# Predicciones y evaluación CNN
y_pred_cnn_scaled = cnn_model.predict(X_test_seq_scaled).flatten()
y_pred_cnn = scaler_y.inverse_transform(y_pred_cnn_scaled.reshape(-1, 1)).flatten()

# Evaluar
cnn_metrics = evaluate_dl_model(y_test_seq, y_pred_cnn, "CNN-1D", history_cnn)

print(f"\nMétricas CNN-1D:")
print(f"RMSE: {cnn_metrics['rmse']:.3f}°C")
print(f"MAE: {cnn_metrics['mae']:.3f}°C")
print(f"R²: {cnn_metrics['r2']:.3f}")
print(f"TSS: {cnn_metrics['tss']:.3f}")

# %% [markdown]
# ## 7. Modelo Híbrido: SARIMA + Red Neuronal

# %%
print("\n" + "="*70)
print("MODELO HÍBRIDO: SARIMA + RED NEURONAL")
print("="*70)

# Este modelo combina:
# 1. SARIMA para capturar la componente lineal/estacional
# 2. Red neuronal para modelar los residuos no lineales

# Cargar predicciones SARIMA del notebook 3 (si existen)
try:
    sarima_predictions = pd.read_csv('../results/predictions/pred_sarimax.csv')
    print("Predicciones SARIMA cargadas")
    
    # Preparar residuos
    residuals = sarima_predictions['tmin_true'] - sarima_predictions['tmin_pred']
    
    # Crear features para predecir residuos
    # Usaremos un subset de features que capturan patrones no lineales
    residual_features = ['temp_range', 'rh', 'pressure', 'wind']
    
    # Preparar datos de residuos
    test_subset = test_data[test_data['station'].isin(sarima_predictions['station'].unique())]
    X_residual = test_subset[residual_features].values
    y_residual = residuals.values
    
    # Escalar
    scaler_residual = StandardScaler()
    X_residual_scaled = scaler_residual.fit_transform(X_residual)
    
    # Modelo simple para residuos
    residual_model = Sequential([
        Input(shape=(len(residual_features),)),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(1)
    ])
    
    residual_model.compile(optimizer=Adam(0.001), loss='mse')
    
    # Entrenar brevemente
    residual_model.fit(
        X_residual_scaled[:int(len(X_residual)*0.8)], 
        y_residual[:int(len(y_residual)*0.8)],
        epochs=20,
        batch_size=64,
        verbose=0
    )
    
    # Predicción final
    residual_pred = residual_model.predict(X_residual_scaled).flatten()
    hybrid_pred = sarima_predictions['tmin_pred'].values + residual_pred
    
    # Evaluar
    hybrid_metrics = evaluate_dl_model(
        sarima_predictions['tmin_true'].values, 
        hybrid_pred, 
        "SARIMA+NN"
    )
    
    print(f"\nMétricas Híbrido SARIMA+NN:")
    print(f"RMSE: {hybrid_metrics['rmse']:.3f}°C")
    print(f"Mejora vs SARIMA solo: calculando...")
    
except:
    print("No se encontraron predicciones SARIMA. Saltando modelo híbrido.")
    hybrid_metrics = None

# %% [markdown]
# ## 8. Comparación de Modelos Deep Learning

# %%
# Recopilar métricas
dl_metrics_list = [mlp_metrics, lstm_metrics, cnn_metrics]
if hybrid_metrics is not None:
    dl_metrics_list.append(hybrid_metrics)

dl_metrics_df = pd.DataFrame(dl_metrics_list)
dl_metrics_df = dl_metrics_df.set_index('model')

print("\n" + "="*70)
print("COMPARACIÓN DE MODELOS DEEP LEARNING")
print("="*70)
print(dl_metrics_df[['rmse', 'mae', 'r2', 'tss']].round(3))

# %%
# Visualización comparativa
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

models = dl_metrics_df.index
metrics_to_plot = ['rmse', 'mae', 'r2', 'precision', 'recall', 'tss']
titles = ['RMSE (°C)', 'MAE (°C)', 'R²', 'Precisión', 'Recall', 'TSS']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum', 'peachpuff']

for i, (metric, title, color) in enumerate(zip(metrics_to_plot, titles, colors)):
    ax = axes[i//3, i%3]
    values = dl_metrics_df[metric].values
    bars = ax.bar(models, values, color=color, alpha=0.8, edgecolor='black')
    
    # Agregar valores
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:.3f}', ha='center', va='bottom')
    
    ax.set_ylabel(title)
    ax.set_title(f'{title} por Modelo')
    ax.grid(True, alpha=0.3, axis='y')
    ax.set_xticklabels(models, rotation=45, ha='right')

plt.suptitle('Comparación de Métricas - Modelos Deep Learning', fontsize=16)
plt.tight_layout()
plt.savefig('../results/figures/dl_models_comparison.png', dpi=300)
plt.show()

# %% [markdown]
# ## 9. Análisis de patrones aprendidos

# %%
# Análisis de activaciones LSTM (ejemplo con una muestra)
print("Analizando patrones aprendidos por LSTM...")

# Tomar una muestra de secuencias
sample_idx = np.random.choice(len(X_test_seq_scaled), 100)
sample_sequences = X_test_seq_scaled[sample_idx]

# Crear modelo intermedio para ver activaciones
# Supongamos que tu lstm_model ya fue entrenado

# El input correcto es el del modelo base
lstm_layer_model = keras.Model(
    inputs=lstm_model.inputs,
    outputs=lstm_model.layers[1].output  # Salida de la primera capa LSTM (ajusta el índice si es necesario)
)


# Obtener activaciones
activations = lstm_layer_model.predict(sample_sequences)

# Visualizar patrones de activación
plt.figure(figsize=(12, 6))

# Promedio de activaciones por unidad LSTM
# Suponiendo activations.shape = (n_muestras, n_timesteps, n_units)
mean_activations = np.mean(activations, axis=(0, 1))  # promedio sobre muestras y timesteps
plt.bar(range(len(mean_activations)), mean_activations)
plt.xlabel('Unidad LSTM')
plt.ylabel('Activación promedio')
plt.title('Activaciones promedio por unidad LSTM')
plt.grid(True, alpha=0.3)


# Heatmap de activaciones para algunas secuencias
# Para mostrar el último timestep de cada muestra
plt.imshow(activations[:, -1, :20].T, aspect='auto', cmap='coolwarm')
plt.colorbar(label='Activación')
plt.xlabel('Secuencia')
plt.ylabel('Unidad LSTM')
plt.title('Patrón de activaciones LSTM')


plt.tight_layout()
plt.savefig('../results/figures/lstm_activation_patterns.png', dpi=300)
plt.show()



# %% [markdown]
# ## 10. Predicción en tiempo real (ejemplo)

# %%
# Función para predicción en tiempo real
def predict_temperature(model, scaler_X, scaler_y, recent_data, model_type='mlp'):
    """
    Realiza predicción con un modelo entrenado
    
    Args:
        model: Modelo entrenado
        scaler_X: Escalador de features
        scaler_y: Escalador de target
        recent_data: Datos recientes (DataFrame o array)
        model_type: 'mlp', 'lstm', o 'cnn'
    
    Returns:
        Predicción de temperatura mínima
    """
    if model_type == 'mlp':
        # Preparar features
        X = recent_data[feature_columns].values
        X_scaled = scaler_X.transform(X)
        
        # Predecir
        y_pred_scaled = model.predict(X_scaled).flatten()
        y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
        
    else:  # lstm o cnn
        # Necesita secuencia de últimos SEQUENCE_LENGTH días
        if len(recent_data) < SEQUENCE_LENGTH:
            raise ValueError(f"Se necesitan al menos {SEQUENCE_LENGTH} días de datos")
        
        # Tomar últimos días
        X = recent_data[feature_columns].values[-SEQUENCE_LENGTH:]
        X = X.reshape(1, SEQUENCE_LENGTH, -1)
        
        # Escalar
        X_scaled = X.copy()
        for i in range(X.shape[2]):
            mean = X[:, :, i].mean()
            std = X[:, :, i].std()
            X_scaled[:, :, i] = (X[:, :, i] - mean) / std
        
        # Predecir
        y_pred_scaled = model.predict(X_scaled).flatten()
        y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
    
    return y_pred[0]

# Ejemplo de uso
print("Ejemplo de predicción en tiempo real:")
sample_station = test_data['station'].iloc[0]
recent_data = test_data[test_data['station'] == sample_station].iloc[:10]

try:
    pred_mlp = predict_temperature(mlp_model, scaler_X, scaler_y, recent_data.iloc[-1:], 'mlp')
    pred_lstm = predict_temperature(lstm_model, scaler_X, scaler_y, recent_data, 'lstm')
    
    print(f"\nEstación: {sample_station}")
    print(f"Temperatura real: {recent_data.iloc[-1]['tmin']:.1f}°C")
    print(f"Predicción MLP: {pred_mlp:.1f}°C")
    print(f"Predicción LSTM: {pred_lstm:.1f}°C")
except Exception as e:
    print(f"Error en predicción: {e}")

# %% [markdown]
# ## 11. Guardar modelos y resultados

# %%
# Guardar modelos
print("Guardando modelos Deep Learning...")

# Guardar arquitecturas y pesos
mlp_model.save('../models/trained/mlp_model.h5')
lstm_model.save('../models/trained/lstm_model.h5')
cnn_model.save('../models/trained/cnn_model.h5')

# Guardar escaladores
joblib.dump({
    'scaler_X': scaler_X,
    'scaler_y': scaler_y,
    'feature_columns': feature_columns,
    'sequence_length': SEQUENCE_LENGTH
}, '../models/trained/dl_scalers.pkl')

# Guardar predicciones
for model_name, predictions, y_true in [
    ('mlp', y_pred_mlp, y_test_mlp),
    ('lstm', y_pred_lstm, y_test_seq),
    ('cnn', y_pred_cnn, y_test_seq)
]:
    # Ajustar longitudes si es necesario
    min_len = min(len(predictions), len(y_true))
    
    pred_df = pd.DataFrame({
        'date': test_info['dates'][:min_len] if model_name == 'mlp' else test_data['date'].iloc[-min_len:],
        'station': test_info['stations'][:min_len] if model_name == 'mlp' else test_data['station'].iloc[-min_len:],
        'tmin_true': y_true[:min_len],
        'tmin_pred': predictions[:min_len],
        'error': y_true[:min_len] - predictions[:min_len]
    })
    
    output_path = f'../results/predictions/pred_{model_name}.csv'
    pred_df.to_csv(output_path, index=False)
    print(f"Predicciones guardadas: {output_path}")

# %%
# Guardar resumen
dl_summary = {
    'models': {
        'mlp': {
            'architecture': 'Feedforward 4 capas [128-64-32-16]',
            'parameters': int(mlp_model.count_params()),
            'metrics': mlp_metrics,
            'training_epochs': len(history_mlp.history['loss'])
        },
        'lstm': {
            'architecture': 'LSTM 2 capas [64-32]',
            'parameters': int(lstm_model.count_params()),
            'metrics': lstm_metrics,
            'training_epochs': len(history_lstm.history['loss']),
            'sequence_length': SEQUENCE_LENGTH
        },
        'cnn1d': {
            'architecture': 'Conv1D 3 capas [64-32-16]',
            'parameters': int(cnn_model.count_params()),
            'metrics': cnn_metrics,
            'training_epochs': len(history_cnn.history['loss']),
            'sequence_length': SEQUENCE_LENGTH
        }
    },
    'best_model': dl_metrics_df['rmse'].idxmin(),
    'best_rmse': float(dl_metrics_df['rmse'].min()),
    'feature_columns': feature_columns,
    'training_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
}

with open('../results/metrics/dl_models_summary.json', 'w') as f:
    json.dump(dl_summary, f, indent=4, default=str)

# %% [markdown]
# ## 12. Conclusiones

print("\n" + "="*70)
print("RESUMEN FINAL - MODELOS DEEP LEARNING")
print("="*70)

best_dl_model = dl_metrics_df['rmse'].idxmin()
print(f"\nMejor modelo DL: {best_dl_model}")
print(f"RMSE: {dl_metrics_df.loc[best_dl_model, 'rmse']:.3f}°C")
print(f"R²: {dl_metrics_df.loc[best_dl_model, 'r2']:.3f}")
print(f"TSS (heladas): {dl_metrics_df.loc[best_dl_model, 'tss']:.3f}")

# Comparación con modelos anteriores
ml_baseline = 1.78  # XGBoost del notebook anterior
statistical_baseline = 2.31  # Prophet

print(f"\nComparación con otros enfoques:")
print(f"vs XGBoost (mejor ML): {(ml_baseline - dl_metrics_df['rmse'].min()) / ml_baseline * 100:.1f}% {'mejor' if dl_metrics_df['rmse'].min() < ml_baseline else 'peor'}")
print(f"vs Prophet (mejor estadístico): {(statistical_baseline - dl_metrics_df['rmse'].min()) / statistical_baseline * 100:.1f}% mejor")

print("\nInsights clave:")
print("1. LSTM captura mejor las dependencias temporales de largo plazo")
print("2. CNN-1D es eficiente para detectar patrones locales en las series")
print("3. MLP es más rápido pero pierde información temporal importante")
print("4. Los modelos DL son competitivos con XGBoost, especialmente LSTM")

print("\nRecomendaciones:")
print("- LSTM para máxima precisión cuando hay suficiente historia")
print("- CNN-1D para balance entre velocidad y precisión")
print("- Considerar ensemble de LSTM + XGBoost para mejor desempeño")
print("- Explorar arquitecturas más profundas o attention mechanisms")

print("\nPróximo paso: Crear ensemble combinando los mejores modelos")

# %% [markdown]
# ## Deep Learning - Conclusiones
# 
# ### Resultados principales:
# - **LSTM** generalmente obtiene el mejor desempeño entre modelos DL
# - Captura dependencias temporales mejor que modelos estáticos
# - Competitivo con XGBoost pero con mayor costo computacional
# 
# ### Ventajas del Deep Learning:
# - Aprende representaciones automáticamente
# - Maneja secuencias temporales naturalmente
# - Escalable a datasets más grandes
# - Potencial para mejoras con más datos
# 
# ### Limitaciones:
# - Requiere más datos para entrenar efectivamente
# - Mayor tiempo de entrenamiento
# - Menos interpretable que modelos tradicionales
# - Sensible a hiperparámetros
# 
# ### Próximos pasos:
# - Implementar modelos de atención (Transformers)
# - Explorar arquitecturas híbridas CNN-LSTM
# - Transfer learning con modelos preentrenados
# - Ensemble con modelos ML para mejor desempeño 
```
